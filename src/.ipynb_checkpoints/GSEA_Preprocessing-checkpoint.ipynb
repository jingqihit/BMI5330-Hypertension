```python
import sys
import os
import numpy as np

gene_symbol_map = {}
with open( '../data/gene_id_names.txt' ) as infile:
    for line in infile:
        geneid = line.strip().split( '\t' )[0]
        genesymbol = line.strip().split( '\t' )[-2]
        genesymbol = genesymbol.strip().strip( '"' ).strip()
        if genesymbol == '':
            continue
        gene_symbol_map[ geneid ] = genesymbol
        print( geneid + '\t' + genesymbol )
print( len( gene_symbol_map ) )
```

```python
start = False
with open( '../data/GSE113439_series_matrix.txt' ) as infile, \
        open( '../data/GSE113439.cls', 'w' ) as clsfile, \
        open( '../data/GSE113439.gct', 'w' ) as outfile:
    groups = []
    samples = []
    for line in infile:
        if line.find( '!Sample_title' ) == 0:
            for sample_title in line.strip().split( "\t" )[1:]:
                if sample_title.find( 'Normal_control' ) >= 0:
                    groups.append( False )
                else:
                    groups.append( True )
        if line.find( '!series_matrix_table_begin' ) == 0:
            line = infile.readline()
            start = True
        
        if not start:
            continue
        if line.find( 'ID_REF' ) >= 0:
            samples = [ a.strip('"') for a in line.strip().split()[1:] ]
            outfile.write( '#1.2\n' )
            outfile.write( str(len( gene_symbol_map )) + ' ' + str(len(samples)) + '\n' )
            outfile.write( 'Name\tDescription\t' + '\t'.join( samples ) + '\n' )
            continue
            
        geneid = line.strip().split( '\t' )[0]
        if geneid not in gene_symbol_map:
            continue
        symbol = gene_symbol_map[ geneid ]
        if symbol.strip() == '':
            continue
        values = '\t'.join( line.strip().split( '\t' )[1:] )
        outfile.write( symbol + '\tdescription\t' + values + '\n' )
        
    clsfile.write( str(len(samples)) + " 2 1\n" )
    clsfile.write( "# Hypertension Normal\n" )
    v = ''
    for f in groups:
        if f:
            v += '0 '
        else:
            v += '1 '
    clsfile.write( v.strip() + '\n' )
        
```
